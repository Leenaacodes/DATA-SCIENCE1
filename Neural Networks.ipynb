```python
from sklearn.preprocessing import LabelEncoder
x = data.iloc[:,0:60]
y = data.iloc[:,-1]

# Encode the target variable 'Y'
encoder = LabelEncoder()
y = encoder.fit_transform(y)
```

```python
#Standardisation
a = StandardScaler()
a.fit(x)
x = a.transform(x)
```

```python
pd.DataFrame(x).describe()
```

#### 2]Model Implementation

```python
!pip install keras
# Create your first MLP in Keras
from keras.models import Sequential
from keras.layers import Dense
import numpy
```

```python
# create model
model = Sequential()
model.add(Dense(12, input_dim=60, activation='relu'))#1st hidden layer
model.add(Dense(8, activation='relu'))# 2nd hidden layer
model.add(Dense(1, activation='sigmoid'))#output layer
```

```python
# Compile model
#we have claassification model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy','precision'])
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=50)

print(f"Shape of x_train: {x_train.shape}")
print(f"Shape of x_test: {x_test.shape}")
print(f"Shape of y_train: {y_train.shape}")
print(f"Shape of y_test: {y_test.shape}")
```

```python
#fit the model
model.fit(x_train, y_train)
```

```python
# Predict probabilities for the test set
predictions_prob = model.predict(x_test)
# Convert probabilities to binary class labels (0 or 1)
predictions = (predictions_prob > 0.5).astype(int).flatten()
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, predictions)
accuracy
```

 This means that approximately 69.05% of the predictions made by your model on the unseen test data were correct. In the context of your binary classification problem (predicting 'R' or 'M' for sonar signals), this indicates that roughly 69.05% of the time, the model correctly identified whether a signal was from a rock or a mine.

#### 3]Hyperparameter Tuning

```python
from keras.layers import Dense,Dropout,Conv2D,MaxPooling2D,Flatten
from keras.optimizers import RMSprop,Adam,Adagrad
from keras.models import Sequential
from keras.callbacks import EarlyStopping
```

```python
!pip install keras-tuner
```

```python
from keras_tuner import RandomSearch
```

```python
def build_model(hp):
  model=Sequential()#hp Hyperparameter

  #providing range for hidden layers
  for i in range (hp.Int('num of layers',2,20)):
    #providing range for number of neurons in hidden layers
    model.add(Dense(units=hp.Int('num of neurons'+str(i),min_value=32,max_value=512),
                    activation='relu'))
    model.add(Dense(1,activation='sigmoid'))#output layer
    model.compile(optimizer=keras.optimizers.Adam(hp.Choice('learning_rate',values=[1e-2,1e-3,1e-4])),
                  loss='binary_crossentropy',metrics=['accuracy'])
    return model
```

```python
tuner=RandomSearch(build_model,objective='val_accuracy',
                   max_trials=4)
```

```python
tuner.search(x_train,y_train,epochs=10,validation_data=(x_test,y_test))
```

This shows the highest validation accuracy found across all trials completed so far. In this case, Trial 4 yielded the best performance
