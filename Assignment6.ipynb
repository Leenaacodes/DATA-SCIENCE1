### Estimation and Confidence Interval

A]

```python
import numpy as np
print_head=[1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]
n=15

```

```python
m=np.mean(print_head)
v=np.var(print_head,ddof=n-1)
sd=v/n
```

```python
from scipy import stats
Confidence_interval=stats.norm.interval(0.99,loc=m,scale=sd/np.sqrt(n))
print(Confidence_interval)
```

CONFIDENCE INTERVAL RESULT: Here we conclude that 99% confidence that true mean of population for durability of print heads before typing characters(in millions) is in between (1.21,1.26) (characters before failure)

Here we are calculating confidence interval for sample of print heads thus for calculating variance we take degree of freedom (n-1)
We use here t-distribution because given sample is n is less than 30 and population standard deviation is unknown

## Hypothesis Testing

```python
#x is no of units produced in week and follows normal distribution
x=600
oper_cost=1000+500*x
n=25
xbar=3050 #sample of mean weekly cost for 25 restraunts
mu=1000+5*x
s=25*5
import numpy as np


```

```python
mu
```

```python
s
```

1]Hypothesis Statement

```python
H0: mu<=4000 # operating cost model equation is accurate
H1: mu>4000 # operating cost model is not accurate
```

2]Calculate Test Statistic

```python
z_cal=(xbar-mu)/(s/np.sqrt(n))
```

```python
z_cal
```

3]Determine the probability and compare

```python
from scipy import stats
```

```python
alpha=0.05
z_crit=stats.norm.ppf(1-(alpha/2))
z_crit
```

```python
p_value = stats.norm.sf(abs(z_cal))
p_value
```

4]Conclusion

```python
if p_value < alpha:
  print("Reject the null hypothesis. There is enough evidence to suggest that the operating cost model is not accurate.")
else:
  print("Fail to reject the null hypothesis. There is not enough evidence to suggest that the operating cost model is not accurate.")
```
