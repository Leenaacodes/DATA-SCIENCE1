### Histogram(applied on continuous numerical column)

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df["Discount Rate (%)"],bins=20,alpha=0.5,facecolor="orange",edgecolor="black")
```

```python
plt.hist(df["Discount Amount"],bins=20,alpha=0.9,facecolor="yellow",edgecolor="black")
```

```python
plt.hist(df["Net Sales Value"],bins=20,alpha=0.5,facecolor="red",edgecolor="black")
```

```python
df.skew(numeric_only=True)
```

Based on the skewness values calculated :-

1.Volume (2.73): This value is significantly positive, indicating a strong right skew. This means there are many instances of low volume sales and fewer instances of high volume sales.

2.Avg Price (1.91): This positive skew suggests that most products have lower average prices, with a smaller number of products having much higher average prices.

3.Total Sales Value (1.53): Similar to Volume and Avg Price, the positive skew indicates that the majority of sales transactions have lower total values, while a few transactions have considerably higher total values.

4.Discount Rate (%) (-1.06): This negative skew suggests a left skew. This means there are more instances of higher discount rates and fewer instances of very low discount rates.

5.Discount Amount (1.91): This positive skew indicates that most discount amounts are smaller, with a few instances of much larger discount amounts.

6.Net Sales Value (1.54): The positive skew here suggests that most net sales values are lower, with a smaller number of sales generating much higher net values.

### Box Plots(applied on continuous variable)

```python
df["Discount Amount"].plot(kind="box")
```

 In this plot, there are several outliers with significantly higher discount amounts than the majority of the data. This suggests that while most discounts are relatively small, there are a few instances where much larger discounts were applied.

```python
df["Discount Rate (%)"].plot(kind="box")
```

The box plot shows that the majority of the "Discount Amount" values are concentrated at the lower end, with a median close to the bottom of the box.


There are numerous outliers extending far above the upper whisker, indicating that a small number of transactions have significantly higher discount amounts compared to the rest of the data. This confirms the positive skew observed earlier.

```python
plt.boxplot(df["Net Sales Value"],vert=False)
```

This box plot shows a different distribution compared to "Discount Amount". The median is closer to the center of the box, suggesting a more symmetrical distribution within the interquartile range.


However, there are outliers below the lower whisker, indicating some instances of unusually low discount rates. This aligns with the negative skew observed earlier.

### Bar Chart (Applied for categorical data)

```python
df.info()
```

```python
df.Date.value_counts().plot(kind="bar")
```

Each date has the same bar height, indicating an equal frequency of observations for each day within the period covered by the data. This suggests that the data collection or recording process was uniform for each day.

```python
df.Day.value_counts().plot(kind="bar",color="red")
```

We observe that maximum sale is on Thursday while other day has equal sale

```python
df.City.unique()
```

```python
df.City.value_counts().plot(kind="bar",color="brown")
```

As the data present sales for one unique city "C". There is only one bar.

```python
df.BU.value_counts().plot(kind="bar",color="brown")
```

Based on the bar chart of the 'BU' (Business Unit) column, we can conclude that the dataset contains an equal number of entries for each of the three business units: Mobiles, FMCG, and Lifestyle.

```python
df.Brand.value_counts().plot(kind="bar",color="green")
```

Here we observe that maximum sale for the Brand Jeera which is more than 80 units while lowest could be observed for Orange and Samsung near to 30 units.

```python
df.Model.value_counts().plot(kind="bar",color="yellow")
```

Based on the bar plot of the 'Model' column, we can conclude that each product model appears an equal number of times in the dataset (specifically 15 times each, as indicated by the bar heights). This suggests a balanced representation of all product models in the data.
