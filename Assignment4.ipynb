```python
import numpy as np
import pandas as pd
bill=pd.read_csv("Billing_Data.csv")
pat=pd.read_csv("Patient_Data.csv")
bill.info()
```

```python
#1.	Load the patient dataset and show summary with info().
pat.info()
```

```python
#2.	Select only the columns relevant for billing: ['PatientID', 'Department', 'Doctor', 'BillAmount'].
#3.	Drop administrative columns like ['ReceptionistID', 'CheckInTime'].
clean_pat=pat.drop(["ReceptionistID","CheckInTime"],axis=1)
clean_pat
```

```python
#4.	Use groupby to find total bill amount per department.
#5.	Remove duplicate patient records based on PatientID.
print(pat.groupby("BillAmount")["Department"].sum())
pat[pat.duplicated()].shape[0]
cleaned_pat=pat.drop_duplicates()
print(cleaned_pat)
```

```python
#6.	Fill missing BillAmount values with the mean bill amount.
cleaned_pat["BillAmount"].fillna(cleaned_pat["BillAmount"].mean(),inplace=True)
cleaned_pat
```

```python
#7.	Merge the billing dataset with patient dataset on PatientID.
merged_data=pd.merge(bill,cleaned_pat,on="PatientID")
merged_data
```

```python
#8. Concatenate an additional DataFrame that contains new patients for the current week (row-wise).
#9. Concatenate new billing category columns like ['InsuranceCovered', 'FinalAmount'] (column-wise).

# Create a new DataFrame with additional patient records
new_patients_data = {
    'PatientID': [106, 107],
    'Name': ['Frank', 'Grace'],
    'Department': ['Pediatrics', 'Oncology'],
    'Doctor': ['Dr. White', 'Dr. Green'],
    'BillAmount': [4800.0, 9200.0],
    'ReceptionistID': [4, 5],
    'CheckInTime': ['2023-01-15 09:30', '2023-01-16 14:00']
}
new_patients_df = pd.DataFrame(new_patients_data)

# Concatenate the new patient data row-wise
updated_pat = pd.concat([cleaned_pat, new_patients_df], ignore_index=True)
display(updated_pat)

# Create new billing columns
new_billing_columns = {
    'InsuranceCovered_new': [2200, 1800, 2700, 3100, 1100, 2500, 1600],
    'FinalAmount_new': [2600, 3800, 4800, 3000, 4500, 2300, 7600]
}
new_billing_df = pd.DataFrame(new_billing_columns)

# Assuming the order of rows in updated_pat matches the order for new_billing_df
final_merged_data = pd.concat([updated_pat, new_billing_df], axis=1)

print("\nDataFrame after adding new billing columns (column-wise):")
display(final_merged_data)
```
