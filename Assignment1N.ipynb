```python
# 1] PRIME NUMBERS
S=int(input("Enter a number: "))
if S>1:
 for i in range(2,S):
  if S%i==0:
    print("Not a prime number")
  else:
    print("It is a prime number")
  break
else:
  print("Please give positive input")
```

```python
#2]Product of Random Numbers
import numpy as np
n1=np.random.randint(1,500)
n2=np.random.randint(1,500)
print(n1,n2)
product=n1*n2
user_product=int(input("Enter your product of two numbers:"))
if user_product==product:
  print("Correct")
else:
  print("Your product is Incorrect")
```

```python
#3]Getting Squares of Odd Number
for i in range(1,100):
  if i%3==0:
    print(i**2)
```

```python
#4]Hospital Billing System
Room_charge=2000
Doctor_fee=1500
lab_charge=300
medicine_charges=int(input("Medicine charges: "))
stay_days=int(input("No of Days patient stayed:"))
lab_tests=int(input("No of lab tests done:"))
total_room_charges=stay_days*Room_charge
lab_charges=lab_charge*lab_tests
Total_bill=total_room_charges+lab_charges+medicine_charges+Doctor_fee
if Total_bill>10000:
  final_bill=Total_bill-(0.1*Total_bill)
  print(final_bill)
else:
  print(Total_bill)
```
