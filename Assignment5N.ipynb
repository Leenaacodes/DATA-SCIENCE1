### checking for duplicated rows and columns to drop

```python
card2.columns[card2.columns.duplicated()]
#checking for any duplicated columns
```

```python
#droping the duplicated rows
card_cleaned=card2.drop_duplicates()
```

```python
card_cleaned = card_cleaned.reset_index(drop=True)
```

## Treating outliers for columns

```python
card_cleaned.boxplot()
```

```python
def outlier_capping(card_cleaned,column):
 Q1=card_cleaned[column].quantile(0.25)
 Q3=card_cleaned[column].quantile(0.75)
 IQR=Q3-Q1
 UE=Q3+1.5*IQR
 LE=Q1-1.5*IQR
 card_cleaned[column]=card_cleaned[column].apply(lambda x:UE if x>UE else LE if x<LE else x)
for col in card_cleaned.select_dtypes(["int","float"]).columns:
 outlier_capping(card_cleaned,col)
```

```python
card_cleaned.boxplot()
```

## 2]Statistical Summary

```python
card.describe()
```


Heart Rate (LB): The average heart rate is around 133 bpm, but there's a wide range from 51 to 214 bpm, indicated by the relatively high standard deviation and the min/max values.
Accelerations (AC), Fetal Movement (FM), Uterine Contractions (UC), Decelerations (DL, DS, DP): These columns have very small mean values and a significant number of zero values (indicated by the 25th and 50th percentiles being zero for some). This suggests that these events (accelerations, movements, contractions, and decelerations) don't occur constantly but rather intermittently.

The maximum values for some of these are much higher than the mean, indicating that when they do occur, they can be quite pronounced.
Variability (ASTV, MSTV, ALTV, MLTV, Width): These columns show measures of variability in the heart rate. There's a wide range of variability observed, with some negative minimum values which might warrant further investigation for data quality.
